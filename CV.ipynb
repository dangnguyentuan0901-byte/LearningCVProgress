Day2: TENSOR MATH

```python
from tensorflow import math
x=tf.constant([-2,3,5,-7])
print(tf.abs(x))
```

```python
x1=tf.constant([3,5,63,1,2,3,1])
x2=tf.constant([2])
print(tf.add(x1,x2))
```

```python
from tensorflow import linalg
x1=tf.constant([[1,2,3],[4,5,6],[7,8,9]],dtype=tf.float32)
x2=tf.constant([[2,3,5,],[3,1,9],[9,2,6]],dtype=tf.float32)
print(tf.expand_dims(x1, axis=0))
print(tf.einsum(('ij,ij -> ij'),x1,x2))
print(x2@tf.linalg.inv(x2))
```

DAY3: TENSORFLOW FUNCION

```python
x1=tf.constant([[12,2,3,9],[3,1,2,3],[5,6,7,4]])
x2=tf.constant([[11,8,12,2],[2,34,4,5],[2,7,8,9]])
print(x1.shape)
x1_exp=tf.expand_dims(x1,axis=1)
print(x1_exp.shape)
x1_exp=tf.squeeze(x1_exp)
print(x1_exp.shape)
x3=tf.concat([x1,x2],axis=0)
print(x3)
x3=tf.pad(x3,[[1,2],[3,4]])
print(x3)
tf.gather(x3,[3,2],axis=0)
```
